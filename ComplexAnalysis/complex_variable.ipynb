## A.1.2 Countur Integrals

Imagine two points, $A$ and $B$, in the two-dimensional $x, y$ plane. Further, imagine that $A$ and $B$ are the two end-points of the curve $C$ in the plane, as shown in Fig. 1. $A$ is the starting end-point and $B$ is the terminating end-point. Now, suppose that $C$ is divided into n parts (or arcs), with the $\mathrm{k}$-th arc having length $\Delta \mathrm{s}_{\mathrm{k}}$ (where $k$ runs from 1 to $n$ ). Each of these arcs has a projection on the $x$ -axis, where $\Delta x_{k}$ is written as the $x$ -axis projection of $\Delta s_{k} .$ In the same way, $\Delta y_{k}$ as the y-axis projection of $\Delta \mathrm{s}_{\mathrm{k}} .$ Again, see Fig. 1 Finally, it is assumed, as $\mathrm{n} \rightarrow \infty$ that $\Delta s_{k} \rightarrow 0,$ that $\Delta x_{k} \rightarrow 0,$ and that $\Delta y_{k} \rightarrow 0,$ for each and every $k$ (that is, the points along $C$ that divide $C$ into $n$ arcs are distributed, loosely speaking, 'uniformly' along $\mathrm{C}$ ).

Continuing, suppose that some function $\mathrm{h}(\mathrm{x}, \mathrm{y})$ is given, such function is defined at every point along $C$. If the two sums $\sum_{k=1}^{n} h\left(x_{k}, y_{k}\right) \Delta x_{k}$ and $\sum_{k=1}^{n} h\left(x_{k}, y_{k}\right) \Delta y_{k}$
where $\left(x_{k}, y_{k}\right)$ is an arbitrary point in the arc $\Delta s_{k}$, then the limiting values [assuming that these two limits exist, <font color='purple'> then it will be seen what that assumption brings </font>] of these sums could be written as:


 

  
$$
\lim _{n \rightarrow \infty} \sum_{k=1}^{n} h\left(x_{k}, y_{k}\right) \Delta x_{k}=\int_{C} h(x, y) d x=I_{x}
$$ [Eq.21]


   
and

$$
\lim _{n \rightarrow \infty} \sum_{k=1}^{n} h\left(x_{k}, y_{k}\right) \Delta y_{k}=\int_{C} h(x, y) d y=I_{y}
$$ [Eq.22]

The value of a line integral depends, in general, on the coordinates of $A$ and $B$, the function $\mathrm{h}(\mathrm{x}, \mathrm{y}),$ and on the specific path $C$ that connects $A$ and $B$.

### A.1.2.1 General Meaning of line- path- and contour integrals

A curve-, line-, path- or contour integral extends the usual definition of an integral to the integration in the complex plane or in a multidimensional space. The term contour integral is typically reserved for line integrals in the complex plane but does not imply integration over a closed contour.

Also note that "Path integral" is a homonym, which on one hand describes integration along a curve and on the other hand is a description in quantum mechanics that generalizes the action principle of classical mechanics.

#### Thinking About and Defining Line Integrals

Let $\mathcal{C}$ be a smove curve in the complex plane and let $f: \mathbb{C} \rightarrow \mathbb{C}$ be a continuous function. We now subdivide the interval $[a, b]$ into $a=t_0<t_1<t_2<\cdots<t_n=b$. So we are left with $n$ intervals of length $\Delta t=(b-a) / n$

<br><br/>
<p align="center"><img src="images/FLrEj.png" style="width:50%"></p>
<center> Fig. XXX$\quad$ XXX. </center>
<br><br/>

From the figure we can easily see that the displacement between two points on $\mathcal{C}$ is $\gamma\left(t_k\right)-\gamma\left(t_{k-1}\right)=\gamma\left(t_{k-1}+\Delta t\right)-\gamma\left(t_{k-1}\right)$
Now we can form the Riemann sum

$$
\lim _{n \rightarrow \infty} \sum_{k=1}^n f\left(\gamma\left(t_k\right)\right)\left[\gamma\left(t_{k-1}+\Delta t\right)-\gamma\left(t_{k-1}\right)\right]=\lim _{n \rightarrow \infty} \sum_{k=1}^n f\left(\gamma\left(t_k\right)\right)\left[\frac{\gamma\left(t_{k-1}+\Delta t\right)-\gamma\left(t_{k-1}\right)}{\Delta t}\right] \Delta t
$$

In the square brackets we have a differential quotient. Recall that $\Delta t=(b-a) / n$, so if we let $n \rightarrow \infty$, the displacement $\Delta t$ tends to zero. This means that if we take the limit, the expression in the square brackets becomes $\gamma^{\prime}(t)$
The integral

$$
\int_a^b f(\gamma(t)) \cdot \gamma^{\prime}(t) d t
$$

is then the limit of this Riemann sum as the lengths of the subdivision intervals approach zero.

### A.1.2.2 Complex Line Integrals 

Line integrals are also called path or contour integrals. Given the ingredients we define the complex lineintegral $\int_\gamma f(z) d z$ by

$$
\int_\gamma f(z) d z:=\int_a^b f(\gamma(t)) \gamma^{\prime}(t) d t .
$$ [Eq. A.1.2.2.1]

understand that the product $f(\gamma(t)) \gamma^{\prime}(t)$ is just a product of complex numbers.
An alternative notation uses $d z=d x+i d y$ to write

$$
\int_\gamma f(z) d z=\int_\gamma(u+i v)(d x+i d y)
$$ [Eq. A.1.2.2.2]

Let's check that Equations A.1.2.2.1 and A.1.2.2.2 are the same. Equation 4.2.2 is really a multivariable calculus expression, so thinking of $\gamma(t)$ as $(x(t), y(t))$ it becomes

$$
\int_\gamma f(z) d z=\int_a^b\left[u(x(t), y(t))+i v(x(t), y(t)]\left(x^{\prime}(t)+i y^{\prime}(t)\right) d t\right.
$$ [Eq. A.1.2.2.3]

but

$$
u(x(t), y(t))+i v(x(t), y(t))=f(\gamma(t))
$$ [Eq. A.1.2.2.4]

and

$$
x^{\prime}(t)+i y^{\prime}(t)=\gamma^{\prime}(t)
$$ [Eq. A.1.2.2.5]

so the right hand side of Equation $4.2 .2$ is

$$
\int_a^b f(\gamma(t)) \gamma^{\prime}(t) d t
$$ [Eq. A.1.2.2.6]

That is, it is exactly the same as the expression in Equation A.1.2.2.1

> __Theorem: Fundamental Theorem of Complex Line Integrals__
>
> If $f(z)$ is a complex analytic function on an open region $A$ and $\gamma$ is a curve in $A$ from $z_0$ to $z_1$ then
>
$$
 \boxed{%
 \int_\gamma f^{\prime}(z) d z=f\left(z_1\right)-f\left(z_0\right) .
 }% end of scope of \boxed
$$
>
> <font color='purple'>
> Proof:
>
> This is an application of the chain rule. We have
>
$$
\frac{d f(\gamma(t))}{d t}=f^{\prime}(\gamma(t)) \gamma^{\prime}(t) .
$$
>
> So
>
$$
\int_\gamma f^{\prime}(z) d z=\int_a^b f^{\prime}(\gamma(t)) \gamma^{\prime}(t) d t=\int_a^b \frac{d f(\gamma(t))}{d t} d t=\left.f(\gamma(t))\right|_a ^b=f\left(z_1\right)-f\left(z_0\right)
$$
>
> Another equivalent way to state the fundamental theorem is: if $f$ has an antiderivative $F$, i.e. $F^{\prime}=f$ then
>
$$
\int_\gamma f(z) d z=F\left(z_1\right)-F\left(z_0\right) .
$$
> </font>

##  A.1.3 Functions of a Complex Variable

Let's write the complex variable $z$ as

$$
z=x+i y
$$ [Eq.31]

where $x$ and $y$ are each real with each varying over the doubly-infinite interval $-\infty$ to $+\infty,$ and $i=\sqrt{-1}$. Geometrically, By interpreting $z$ as a point in an infinite, two-dimensional plane (called the complex plane) with $x$ measured along a horizontal axis and $y$ measured along a vertical axis. And considering a complex function of the complex variable $z$ as

$$
\mathrm{f}(\mathrm{z})=\mathrm{f}(\mathrm{x}+i \mathrm{y})=\mathrm{u}(\mathrm{x}, \mathrm{y})+i \mathrm{v}(\mathrm{x}, \mathrm{y})
$$ [Eq.32]

where $u$ and $v$ are each real-valued functions of the two real-valued variables $x$ and $y$. 

-----------------------------------------------------------------
> __Example__
>
> If $f(z)=z^{2}=(x+i y)^{2}=x^{2}-y^{2}+i 2 x y$ then,
>
> in this case, $u=x^{2}-y^{2}$ and $v=2 x y$. In $x, y$ notation, It is said to be working in rectangular (or Cartesian) coordinates.
-----------------------------------------------------------------

It is often convenient to work in polar coordinates, which means the complex variable $z$ is written as

$$
z=r e^{i \theta}
$$ [Eq.33]

where $r$ and $\theta$ are each real: $r$ is the radial distance from the origin of the coordinate system of the complex plane to the point $z$ (and so $0 \leq r<\infty$ ), and $\theta$ is the angle of the radius vector (of length $\mathrm{r}$ ) measured counter-clockwise from the positive horizontal x-axis to the radius vector (and so it is $0 \leq \theta<2 \pi$ usually taken). Note, carefully, however, that $\theta$ is not uniquely determined, as it could be added (or subtract) any multiple of $2 \pi$ from $\theta$ and still be talking about the same point in the complex plane.

From Euler's fabulous formula, it is obtained from [Eq.33] that

$$
z=r\{\cos (\theta)+i \sin (\theta)\}
$$ [Eq.34]

-----------------------------------------------------------------
> __Example__
>
> If $f(z)=z^{2}$ then
> 
$$
f(z)=\left(r e^{i \theta}\right)^{2}=r^{2}\{\cos (\theta)+i \sin (\theta)\}^{2}
$$
> 
> or, expanding both sides of the last equality,
> 
$$ 
r^{2} e^{i 2 \theta}=r^{2}\{\cos (2 \theta)+i \sin (2 \theta)\}=r^{2}\left\{\cos ^{2}(\theta)-\sin ^{2}(\theta)+i 2 \cos (\theta) \sin (\theta)\right\}
$$
> 
> since the real and imaginary parts of the expressions in the last equality must be separately equal, it is concluded that $\cos (2 \theta)=\cos ^{2}(\theta)-\sin ^{2}(\theta)$ as well as $\sin (2 \theta)= 2 \cos (\theta) \sin (\theta)$. These two formulas are the well-known double-angle formulas from trigonometry, and so already it is a nice illustration of the powerful capability of complex functions to do useful work for calculi.

f(z)=\frac{e^{\frac{l}{z}\left(z+\frac{1}{z}\right)}}{z}
$$
> 
> which will be integrated counter-clockwise once around the unit circle. That is, it will be computed
> 
$$
\oint_{C} \frac{e^{\frac{1}{2}\left(z+\frac{1}{2}\right)}}{z} d z
$$
> 
> where $C$ is the circle $z=e^{i \theta}$. [The circle on the integral sign is there simply to emphasize that work is being performed with a closed line integral.].
> 
> The reason for the $z$ in the denominator of the integrand is because $d z=i e^{i \theta} d \theta$ and it is needed an $\mathrm{e}^{i \theta}$ in the denominator to cancel the $\mathrm{e}^{i \theta}$ in $\mathrm{dz}$. So,
> 
$$
\oint_{C} \frac{\mathrm{e}^{\frac{1}{2}\left(\mathrm{z}+\frac{1}{2}\right)}}{\mathrm{z}} \mathrm{d} \mathrm{z}=\int_{0}^{2 \pi} \frac{\mathrm{e}^{\cos (\theta)}}{\mathrm{e}^{i \theta}} i \mathrm{e}^{i \theta} \mathrm{d} \theta=i \int_{0}^{2 \pi} \mathrm{e}^{\cos (\theta)} \mathrm{d} \theta
$$ [Eq.35]
> 
> That is, the contour integral at the left in [Eq.35] is the integral it were the interest on after multiplied by $i$. 
> 
> To directly calculate the contour integral, let's start by expanding the exponential in the left-most integral in a power series. That is,
> 
$$
\oint_{\mathrm{C}} \frac{\mathrm{e}^{\frac{1}{2}\left(\mathrm{z}+\frac{1}{2}\right)}}{\mathrm{z}} \mathrm{d} z=\oint_{\mathrm{C}} \frac{1}{\mathrm{z}} \sum_{\mathrm{n}=0}^{\infty} \frac{\left\{\frac{1}{2}\left(\mathrm{z}+\frac{1}{\mathrm{z}}\right)\right\}^{\mathrm{n}}}{\mathrm{n} !} \mathrm{d} \mathrm{z}=\oint_{\mathrm{C}} \frac{1}{\mathrm{z}} \sum_{\mathrm{n}=0}^{\infty} \frac{1}{2^{\mathrm{n}} \mathrm{n} !}\left(\mathrm{z}+\frac{1}{\mathrm{z}}\right)^{\mathrm{n}} \mathrm{d} z
$$
> 
> Using the binomial theorem to write
> 
$$
\left(z+\frac{1}{z}\right)^{n}=\sum_{k=0}^{n}\left(\begin{array}{l}
n \\
k
\end{array}\right) z^{k}\left(\frac{1}{z}\right)^{n-k}=\sum_{k=0}^{n}\left(\begin{array}{l}
n \\
k
\end{array}\right) z^{k} \frac{1}{z^{n-k}}=\sum_{k=0}^{n}\left(\begin{array}{l}
n \\
k
\end{array}\right) \frac{z^{2 k}}{z^{n}}
$$
> 
> it is obtained
> 
$$
\begin{aligned}
\oint_{C} \frac{e^{\frac{1}{2}\left(z+\frac{1}{z}\right)}}{z} d z &=\oint_{C} \frac{1}{z} \sum_{n=0}^{\infty} \frac{1}{2^{n} n !}\left\{\sum_{k=0}^{n}\left(\begin{array}{c}
n \\
k
\end{array}\right) \frac{z^{2 k}}{z^{n}}\right\} d z \\
&=\oint_{C} \sum_{n=0}^{\infty} \frac{1}{2^{n} n !}\left\{\sum_{k=0}^{n}\left(\begin{array}{c}
n \\
k
\end{array}\right) z^{2 k-n-1}\right\} d z \\
&=\sum_{n=0}^{\infty} \frac{1}{2^{n} n !}\left\{\sum_{k=0}^{n}\left(\begin{array}{c}
n \\
k
\end{array}\right) \oint_{C} z^{2 k-n-1}\right\} d z
\end{aligned}
$$
> 
> Now, concentrate on that last integral, where it is replaced $z$ with $e^{i \theta}$ and $d z$ with $i \mathrm{e}^{i \theta} \mathrm{d} \theta$
> 
$$
\oint_{\mathrm{C}} \mathrm{z}^{2 \mathrm{k}-\mathrm{n}-1} \mathrm{d} z=\int_{0}^{2 \pi}\left(\mathrm{e}^{\mathrm{i} \theta}\right)^{2 \mathrm{k}-\mathrm{n}-1} i \mathrm{e}^{i \theta} \mathrm{d} \theta=i \int_{0}^{2 \pi} \mathrm{e}^{i(2 \mathrm{k}-\mathrm{n}) \theta} \mathrm{d} \theta=\begin{array}{c}
2 \pi i \text { if } 2 \mathrm{k}-\mathrm{n}=0 \\
0 \text { if } 2 \mathrm{k}-\mathrm{n} \neq 0
\end{array}
$$
> 
> This is remarkable! Every one of these integrals on the right vanishes as $n$ and $k$ run through their values except for those cases where $k=\frac{n}{2}$. This has a profound implication, as then $\mathrm{k}$ can be an integer (which of course it is) only if $\mathrm{n}$ is even. For all odd values of $n$ the integrals vanish, and in the cases of $n$ even they vanish, too, if $\mathrm{k} \neq \frac{\mathrm{n}}{2} .$ It could be possible to include all the integrals that don't vanish with the simple trick of writing $n=2 \mathrm{m},$ where $\mathrm{m}=0,1,2,3, \ldots,$ and so to have
> 
$$
\begin{aligned}
\oint_{C} \frac{\mathrm{e}^{\frac{1}{2}\left(\mathrm{z}+\frac{1}{2}\right)}}{\mathrm{z}} \mathrm{d} \mathrm{z} &=\sum_{\mathrm{m}=0}^{\infty} \frac{1}{2^{2 \mathrm{m}}(2 \mathrm{m} !)}\left(\begin{array}{c}
2 \mathrm{m} \\
\mathrm{m}
\end{array}\right) 2 \pi \mathrm{i}=2 \pi \mathrm{i} \sum_{\mathrm{m}=0}^{\infty} \frac{1}{2^{2 \mathrm{m}}(2 \mathrm{m} !)} \frac{(2 \mathrm{m} !)}{\mathrm{m} ! \mathrm{m} !} \\
&=2 \pi \mathrm{i} \sum_{\mathrm{m}=0}^{\infty} \frac{1}{2^{2 \mathrm{m}}(\mathrm{m} !)^{2}}
\end{aligned}
$$
> 
> From [Eq.35] it can now be written
> 
$$
2 \pi \mathrm{i} \sum_{\mathrm{m}=0}^{\infty} \frac{1}{2^{2 \mathrm{m}}(\mathrm{m} !)^{2}}=i \int_{0}^{2 \pi} \mathrm{e}^{\cos (\theta)} \mathrm{d} \theta
$$
> 
> or, cancelling the $i$'s, obtaining then, the pursuited answer, as:
> 
$$
\int_{0}^{2 \pi} e^{\cos (\theta)} d \theta=2 \pi \sum_{m=0}^{\infty} \frac{1}{2^{2 m}(m !)^{2}}
$$ [Eq.36]
> 
> The terms in the series on the right decrease very rapidly and so the series quickly converges. Using just the first four terms the sum is $2 \pi\left(1+\frac{1}{4}+\frac{1}{64}+\frac{1}{2,304}\right)=7.95488$ and MATLAB/Python codes agrees.

## A.1.4 The Cauchy-Riemann Equations and Analytic Functions

Complex function theory really starts with the study of what it means to talk of the derivative of $\mathrm{f}(\mathrm{z})$. In real function theory, the derivative of $\mathrm{g}(\mathrm{x})$ at $\mathrm{x}=\mathrm{x}_0$ is defined as

$$
\left.\frac{\mathrm{dg}}{\mathrm{dx}}\right|_{\mathrm{x} \rightarrow \mathrm{x}_0}=\lim _{\Delta \mathrm{x} \rightarrow 0} \frac{\mathrm{g}\left(\mathrm{x}_0+\Delta \mathrm{x}\right)-\mathrm{g}\left(\mathrm{x}_0\right)}{\Delta \mathrm{x}}=\mathrm{g}^{\prime}\left(\mathrm{x}_0\right) .
$$

We do almost the same thing with a complex function. Indeed, the formal definition for the derivative of a complex $f(z)$ at $z=z_0$ is

$$
\left.\frac{\mathrm{df}}{\mathrm{dx}}\right|_{\mathrm{z}-\mathrm{z}_0}=\lim _{\Delta \mathrm{z} \rightarrow 0} \frac{\mathrm{f}\left(\mathrm{z}_0+\Delta \mathrm{z}\right)-\mathrm{f}\left(\mathrm{z}_0\right)}{\Delta \mathrm{z}}=\mathrm{f}^{\prime}\left(\mathrm{z}_0\right) .
$$

The vanishing of $\Delta z=\Delta x+i \Delta y$ is, however, not quite as straightforward as it is in the case of a real variable. In that simpler case, where we let $\Delta x \rightarrow 0$ to calculate $\mathrm{g}^{\prime}\left(\mathrm{x}_0\right), \Delta \mathrm{x}$ only has to vanish along the one-dimensional real axis. That is, $\Delta \mathrm{x}$ can shrink to zero in just two ways: either from the left of $x_0$ or from the right of $x_0$. But in the complex case we must take into account that, since $z_0$ is a point in the complex, two-dimensional plane, then $\Delta z$ can shrink to zero in an infinity of different ways (from the left of $\mathrm{z}_0$, from the right of $\mathrm{z}_0$, from below $\mathrm{z}_0$, from above $z_0$ or, indeed, from any direction of the compass). So, just how does $\Delta z \rightarrow 0$ ?
Mathematicians consider the most condition-free definition possible for the derivative to be the best definition, and so their answer to our question is: we want $f^{\prime}\left(z_0\right)$ to be the same independent of how $\Delta z \rightarrow 0$. To have this be the case, as you might suspect, comes with a price. If $f=u+i v$ then the price for a derivative at you might suspect, comes with a price. If $\mathrm{f}=\mathrm{u}+\mathrm{iv}$ then the price for a derivative at $\mathrm{z}=\mathrm{z}_0$ that doesn't depend on the precise nature of how $\Delta \mathrm{z} \rightarrow 0$ is that $\mathrm{u}$ and $\mathrm{v}$ $\mathrm{c}=\mathrm{z}_0$ that doesn't depend on the precise nature of how $\Delta \mathrm{z} \rightarrow 0$ is that $\mathrm{u}$ and $\mathrm{V}$ these conditions are satisfied at $\mathrm{z}=\mathrm{z}_0$ and at all points in a region (domain or neighborhood are terms that are also used) surrounding $\mathrm{z}_0$, then we say that $\mathrm{f}(\mathrm{z})$ is an analyric function in that region (not to be confused with the analytic signal from radio theory that we encountered in the previous chapter).

The conditions for $\mathrm{f}(\mathrm{z})$ to be analytic are called the Cauchy-Riemann (C-R) equations, which are actually pretty easy to state: at $\mathrm{z}=\mathrm{z}_0$ it must be true that

$$
\frac{\partial u}{\partial x}=\frac{\partial v}{\partial y}
$$

and

$$
\frac{\partial u}{\partial y}=-\frac{\partial v}{\partial x}
$$

Using the general form of the binomial theorem, the integrand expands to

$$
\left(1-k^{2} \sin ^{2} \theta\right)^{-\frac{1}{2}}=\sum_{n=0}^{\infty}\left(\begin{array}{c}
-\frac{1}{2} \\
n
\end{array}\right)(-1)^{n} k^{2 n} \sin ^{2 n} \theta
$$

where

$$
\left(\begin{array}{c}
-\frac{1}{2} \\
n
\end{array}\right)=\frac{\left(-\frac{1}{2}\right)\left(-\frac{1}{2}-1\right)\left(-\frac{1}{2}-2\right) \ldots\left(-\frac{1}{2}-(n-1)\right)}{n !}
$$

which simplifies to

$$
(-1)^{n} \frac{1 \cdot 3 \cdot 5 \ldots(2 n-1)}{2^{n} n !}=(-1)^{n} \frac{(2 n) !}{\left(2^{n} n !\right)^{2}}
$$

It follows that to have

$$
K(k)=\sum_{n=0}^{\infty} \frac{(2 n) !}{\left(2^{n} n !\right)^{2}}\left(\int_{0}^{\frac{\pi}{2}} \sin ^{2 n} \theta d \theta\right) k^{2 n}
$$

so the remaining works left with computing the integral of $\sin ^{2 n} \theta$. This is straightforward to compute using Euler's formula, which gives

$$
\int_{0}^{\frac{\pi}{2}} \sin ^{2 n} \theta d \theta=\int_{0}^{\frac{\pi}{2}}\left(\frac{e^{i \theta}-e^{-i \theta}}{2 i}\right)^{2 n} d \theta

Using a second application of the binomial theorem, the $k^{t h}$ term is

\frac{(-1)^{n+k}}{4^{n}}\left(\begin{array}{c}

\end{array}\right) \int_{0}^{\frac{\pi}{2}} e^{i(2 n-2 k) \theta} d \theta

Since it is known the answer is real, the imaginary part of this integral can be ignored and just focus on its real part. The integral of the real part is

vanishes unless $k=n$ by symmetry considerations, so $k=n$ is the only relevant term and from this, it is gotten

n
\end{array}\right) \int_{0}^{\frac{\pi}{2}} 1 d \theta

where the integral is just $\frac{\pi}{2} .$ This gives the final answer

### A.1.13.n Kepler's laws of planetary motion

### First law

> The orbit of every planet is an ellipse with the Sun at one of the two foci.

Mathematically, an ellipse can be represented by the formula:

$$
r=\frac{p}{1+\varepsilon \cos \theta},
$$

where $p$ is the semi-latus rectum, $\varepsilon$ is the eccentricity of the ellipse, $r$ is the distance from the Sun to the planet, and $\theta$ is the angle to the planet's current position from its closest approach, as seen from the Sun. So $(r, \theta)$ are polar coordinates.

For an ellipse $0<\varepsilon<1$; in the limiting case $\varepsilon=0$, the orbit is a circle with the Sun at the centre (i.e. where there is zero eccentricity).

At $\theta=0^{\circ}$, perihelion, the distance is minimum

$$
r_{\min }=\frac{p}{1+\varepsilon}
$$

At $\theta=90^{\circ}$ and at $\theta=270^{\circ}$ the distance is equal to $p$.

At $\theta=180^{\circ}$, aphelion, the distance is maximum (by definition, aphelion is - invariably - perihelion plus $\left.180^{\circ}\right)$

$$
r_{\max }=\frac{p}{1-\varepsilon}
$$

The semi-major axis $a$ is the arithmetic mean between $r_{\text {min }}$ and $r_{\text {max }}$ :

$$
\begin{aligned}
r_{\max }-a &=a-r_{\min } \\
a &=\frac{p}{1-\varepsilon^2}
\end{aligned}
$$

The semi-minor axis $b$ is the geometric mean between $r_{\min }$ and $r_{\max }$ :

$$
\begin{aligned}
\frac{r_{\max }}{b} &=\frac{b}{r_{\min }} \\
b &=\frac{p}{\sqrt{1-\varepsilon^2}}
\end{aligned}
$$

The semi-latus rectum $p$ is the harmonic mean between $r_{\min }$ and $r_{\text {max }}$ :

$$
\begin{aligned}
\frac{1}{r_{\min }}-\frac{1}{p} &=\frac{1}{p}-\frac{1}{r_{\max }} \\
p a &=r_{\max } r_{\min }=b^2
\end{aligned}
$$

The eccentricity $\varepsilon$ is the coefficient of variation between $r_{\min }$ and $r_{\text {max }}$ :

$$
\varepsilon=\frac{r_{\max }-r_{\min }}{r_{\max }+r_{\min }} .
$$

The area of the ellipse is

$$
A=\pi a b .
$$

The special case of a circle is $\varepsilon=0$, resulting in $r=p=r_{\min }=r_{\max }=a=b$ and $A=\pi r^2$.

### Defining the Ellipse's Equation in the Complex Plane

#### Method 1.
For $x, y, h, k, a, b \in \mathbb{R}$ such that $a, b \neq 0$, we define a real ellipse as

$$
\frac{(x-h)^2}{a^2}+\frac{(y-k)^2}{b^2}=1
$$

Let's consider the restriction: $z=\frac{x}{a}+i \frac{y}{b}$ and $z_0=\frac{h}{a}+i \frac{k}{b}$. If we expand the equation for an ellipse, we have

$$
\frac{x^2}{a^2}+\frac{y^2}{b^2}+\frac{h^2}{a^2}+\frac{k^2}{b^2}-\frac{2 x h}{a^2}-\frac{2 y k}{b^2}=1
$$

Notice that $|z|^2=\frac{x^2}{a^2}+\frac{y^2}{b^2}$ and $\left|z_0\right|^2=\frac{h^2}{a^2}+\frac{h^2}{b^2}$. Now, let's write the ellipse as

$$
|z|^2+\left|z_0\right|^2-\frac{2 x h}{a^2}-\frac{2 y k}{b^2}+\frac{y h}{a b} i-\frac{y h}{a b} i+\frac{x k}{a b} i-\frac{x k}{a b} i=|z|^2+\left|z_0\right|^2-\bar{z} z_0-z \bar{z}_0=1 .
$$

Thus, the equation of an ellipse in the complex plane is

$$
\left(z-z_0\right)\left(\bar{z}-\bar{z}_0\right)=\left|z-z_0\right|^2=1 \Rightarrow\left|z-z_0\right|=1
$$

where $z$ and $z_0$ are defined above.

#### Method 2.

You can start with the parametrization

$$
\begin{aligned}
&x(s)=a \cos s \\
&y(s)=b \sin s
\end{aligned}
$$

and write

$$
\begin{aligned}
&z(s)=x(s)+i y(s)=a(\cos s+i \sin s)+(b-a)(i \sin s) \\
&=a e^{i s}+\frac{b-a}{2}\left(e^{i s}-e^{-i s}\right) \\
&=\left(a+\frac{b-a}{2}\right) e^{i s}-\left(\frac{b-a}{2}\right) e^{-i s} \\
&=A_{+} e^{i s}+A e^{-i s} \\
&\text { with } A_{\pm}=\frac{a \pm b}{2} \in \mathbb{R}
\end{aligned}
$$

There are several virtues to this representation. If you want the ellipse to be centered around a point $z_0=x_0+i y_0$, just add it:

$$
z(s) \rightarrow z_0+A_{+} e^{i s}+A_{-} e^{-i s}
$$

If you want the ellipse to be rotated by an angle $\psi$, just multiply:

$$
z(s) \rightarrow e^{i \psi}\left(A_{+} e^{i s}+A_{-} e^{-i s}\right)=A_{+} e^{i(s+\psi)}+A_{-} e^{-i(s-\psi)}
$$

### Second law

> A line joining a planet and the Sun sweeps out equal areas during equal intervals of time.

The orbital radius and angular velocity of the planet in the elliptical orbit will vary. This is shown in the animation: the planet travels faster when closer to the Sun, then slower when farther from the Sun. Kepler's second law states that the blue sector has constant area.

In a small time $d t$ the planet sweeps out a small triangle having base line $r$ and height $r d \theta$ and area $d A=\frac{1}{2} \cdot r \cdot r d \theta$, so the constant areal velocity is

$$
\frac{d A}{d t}=\frac{r^2}{2} \frac{d \theta}{d t} .
$$

The area enclosed by the elliptical orbit is $\pi a b$. So the period $T$ satisfies

$$
T \cdot \frac{r^2}{2} \frac{d \theta}{d t}=\pi a b
$$

and the mean motion of the planet around the Sun

$$
n=\frac{2 \pi}{T}
$$

satisfies

$$
r^2 d \theta=(a b n) dt .
$$

And so,

$$
\frac{d A}{d t}=\frac{a b n}{2}=\frac{\pi a b}{T}
$$

### Kepler's Second law with Complex Variable

Imagine a point mass $m$ moving in a plane along the path given by [Eq. X37],

$$
z(t)=r(t) e^{i \theta(t)}
$$ [Eq. X37]

where now $z, r$ and $\theta$ are specifically indicated to be functions of time (t). The motion of $\mathrm{m}$ is due entirely to a force acting along the line connecting the mass to the source of the force: the classic example of this situation is the Earth (the 'point' mass) moving under the influence of the gravitational field of the Sun (which it'll be taken as being at the origin of the $x-y$ coordinate system). The attractive force on the Earth is, of course, always directed radially inward towards the Sun.

If the magnitude of the force are written on $\mathrm{m}$ as $\mathrm{f}(\mathrm{r}, \theta)$, Newton's famous second law of motion ('force is mass times acceleration') says

$$
\mathrm{f}(\mathrm{r}, \theta) \mathrm{e}^{i \theta}=\mathrm{m} \frac{\mathrm{d}^2 \mathrm{z}}{\mathrm{dt}^2} .
$$ [Eq. X38]

From [Eq. X37] it has

$$
\frac{\mathrm{dz}}{\mathrm{dt}}=\frac{\mathrm{dr}}{\mathrm{dt}} \mathrm{e}^{i \theta}+i \mathrm{ir} \frac{\mathrm{d} \theta}{\mathrm{dt}} \mathrm{e}^{i \theta}
$$

and so

$$
\frac{\mathrm{d}^2 \mathrm{z}}{\mathrm{dt}^2}=\frac{\mathrm{d}^2 \mathrm{r}}{\mathrm{dt}^2} \mathrm{e}^{i \theta}+\frac{\mathrm{dr}}{\mathrm{dt}} i \frac{\mathrm{d} \theta}{\mathrm{dt}} \mathrm{e}^{i \theta}+i\left[\frac{\mathrm{dr}}{\mathrm{dt}} \frac{\mathrm{d} \theta}{\mathrm{dt}} \mathrm{e}^{i \theta}+\mathrm{r} \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2} \mathrm{e}^{i \theta}+\mathrm{r} \frac{\mathrm{d} \theta}{\mathrm{dt}} i \frac{\mathrm{d} \theta}{\mathrm{dt}} \mathrm{e}^{i \theta}\right]
$$

or,

$$
\frac{\mathrm{d}^2 \mathrm{z}}{\mathrm{dt}^2}=\left[\frac{\mathrm{d}^2 \mathrm{r}}{\mathrm{dt}^2}-\mathrm{r}\left(\frac{\mathrm{d} \theta}{\mathrm{dt}}\right)^2\right] \mathrm{e}^{3 \theta}+i\left[2 \frac{\mathrm{dr}}{\mathrm{dt}} \frac{\mathrm{d} \theta}{\mathrm{dt}}+\mathrm{r} \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2}\right] \mathrm{e}^{3 \theta} .
$$ [Eq. X39]

Using [Eq. X39] in [Eq. X38] and cancelling all the $\mathrm{e}^{i t}$ (which are never zero), it arrives at 

$$
\mathrm{f}(\mathrm{r}, \theta)=\mathrm{m}\left[\frac{\mathrm{d}^2 \mathrm{r}}{\mathrm{dt}^2}-\mathrm{r}\left(\frac{\mathrm{d} \theta}{\mathrm{dt}}\right)^2\right]+i \mathrm{~m}\left[2 \frac{\mathrm{dr}}{\mathrm{dt}} \frac{\mathrm{d} \theta}{\mathrm{dt}}+\mathrm{r} \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2}\right] .
$$

Equating real and imaginary parts of this last expression gives the famous differential equations of motion in a radial force field:

$$
f(r, \theta)=m\left[\frac{d^2 r}{d t^2}-r\left(\frac{d \theta}{d t}\right)^2\right]
$$ [Eq. X40]

and

$$
2 \frac{\mathrm{dr}}{\mathrm{dt}} \frac{\mathrm{d} \theta}{\mathrm{dt}}+\mathrm{r} \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2}=0 .
$$ [Eq. X41]

Interestingly, the result in [Eq. 41] was implicitly known long before Newton. Mathematics alone tells us that

$$
\frac{\mathrm{d}}{\mathrm{dt}}\left\{\mathrm{r}^2 \frac{\mathrm{d} \theta}{\mathrm{dt}}\right\}=2 \mathrm{r}\left(\frac{\mathrm{dr}}{\mathrm{dt}}\right) \frac{\mathrm{d} \theta}{\mathrm{dt}}+\mathrm{r}^2 \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2}=\mathrm{r}\left[2\left(\frac{\mathrm{dr}}{\mathrm{dt}}\right) \frac{\mathrm{d} \theta}{\mathrm{dt}}+\mathrm{r} \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2}\right]
$$

and, since the expression in the square brackets is zero by [Eq. X41] , it has

$$
\frac{d}{d t}\left\{r^2 \frac{d \theta}{d t}\right\}=0 .
$$

$$
\mathrm{r}^2 \frac{d \theta}{\mathrm{dt}}=\mathrm{C}
$$ [Eq. X42]

where $\mathrm{C}$ is a constant. This result has a historically important physical interpretation in the theory of planetary motion.

### Third law

> The ratio of the square of an object's orbital period with the cube of the semi-major axis of its orbit is the same for all objects orbiting the same primary.

This captures the relationship between the distance of planets from the Sun, and their orbital periods.
Kepler enunciated in 1619, this third law in a laborious attempt to determine what he viewed as the "music of the spheres" according to precise laws, and express it in terms of musical notation. It was therefore known as the harmonic law.

Using Newton's law of gravitation (published 1687), this relation can be found in the case of a circular orbit by setting the centripetal force equal to the gravitational force:
$$
m r \omega^2=G \frac{m M}{r^2}
$$
Then, expressing the angular velocity $\omega$ in terms of the orbital period $T$ and then rearranging, results in Kepler's Third Law:
$$
m r\left(\frac{2 \pi}{T}\right)^2=G \frac{m M}{r^2} \rightarrow T^2=\left(\frac{4 \pi^2}{G M}\right) r^3 \rightarrow T^2 \propto r^3
$$
A more detailed derivation can be done with general elliptical orbits, instead of circles, as well as orbiting the center of mass, instead of just the large mass. This results in replacing a circular radius, $r$, with the semi-major axis, $a$, of the elliptical relative motion of one mass relative to the other, as well as replacing the large mass $M$ with $M+m$. However, with planet masses being so much smaller than the Sun, this correction is often ignored. The full corresponding formula is:
$\frac{a^3}{T^2}=\frac{G(M+m)}{4 \pi^2} \approx \frac{G M}{4 \pi^2} \approx 7.496 \times 10^{-6} \frac{\mathrm{AU}^3}{\text { days }^2}$ is constant
where $M$ is the mass of the Sun, $m$ is the mass of the planet, $G$ is the gravitational constant, $T$ is the orbital period and $a$ is the elliptical semi-major axis, and $\mathrm{AU}$ is the astronomical unit, the average distance from earth to the sun.

### What is it about the Navier-Stokes equations that make them so challenging to find solutions? 
### What makes them different from other differential equations that there are solutions for?

The Navier Stokes equations are a set of equations that capture the interrelationships among the various parameters in fluid mechanics, such as velocity, pressure, density, temperature, etc.

https://en.wikipedia.org/wiki/Navier%E2%80%93Stokes_equations

Actually, there various sets of these equations depending on the simplifications that are justified for various flow situations. For example, for low speed flows, the changes of pressure are small enough that they do not change the density of the fluid enough to account for. So we simplify the equations by assuming the density is constant. This is called incompressible flow. When the flow is VERY slow, the effects of inertia can also be ignored. In that case, called creeping flow, you can drop the time derivatives. In many cases, the changes of temperature associated with the flow are not important and then we don't need the energy equation, which is one of the equations in the full set.

The equations are coupled second order partial differential equations. Three of the equations are related to the changes in momentum of the fluid and how those changes go along with pressure and viscous stresses. These are one for each of the coordinate in a 3D world $-x, y, z$ components of velocity. Here's one version of the three equations dealing with momentum:

Cartesian coordinates

From the general form of the Navier-Stokes, with the velocity vector expanded as $\mathbf{u}=\left(u_x, u_y, u_z\right)$, sometimes respectively named $u_n, v, w$, we may write the vector equation explicitly,

$$
\begin{gathered}
x: \rho\left(\partial_t u_x+u_z \partial_x u_z+u_y \partial_y u_z+u_z \partial_z u_z\right) \\
=-\partial_z p+\mu\left(\partial_z^2 u_z+\partial_y^2 u_z+\partial_z^2 u_x\right)+\frac{1}{3} \mu \partial_x\left(\partial_x u_z+\partial_y u_y+\partial_z u_z\right)+\rho g_x \\
y: \rho\left(\partial_t u_y+u_x \partial_x u_y+u_y \partial_y u_y+u_z \partial_z u_y\right) \\
=-\partial_y p+\mu\left(\partial_x^2 u_y+\partial_y^2 u_y+\partial_z^2 u_y\right)+\frac{1}{3} \mu \partial_y\left(\partial_x u_x+\partial_y u_y+\partial_z u_z\right)+\rho g_y \\
z: \rho\left(\partial_t u_x+u_z \partial_x u_z+u_y \partial_y u_z+u_z \partial_z u_z\right) \\
=-\partial_z p+\mu\left(\partial_x^2 u_z+\partial_y^2 u_z+\partial_z^2 u_z\right)+\frac{1}{3} \mu \partial_z\left(\partial_z u_x+\partial_y u_y+\partial_z u_z\right)+\rho g_z
\end{gathered}
$$

Note that gravity has been accounted for as a body force, and the values of $g_x . g_y$. $g_z$ will depend on the orientation of gravity with respect to the chosen set of coordinates.

A fourth equation, called the continuity equation, deals with tracking how the density of the fluid goes along with the moving fluid. When the fluid can be treated as constant density, this equation effectively enforces conservation of matter.

The continuity equation reads:

$$
\partial_t \rho+\partial_x\left(\rho u_x\right)+\partial_y\left(\rho u_y\right)+\partial_z\left(\rho u_z\right)=0
$$

The fifth main equation is the energy equation. It deals with various interactions within the fluid that have changes of energy. This can take the form of an equation of state from thermodynamics. Here's about the simplest equation of state where there is some variation the density and temperature. This is the ideal gas law:

$$
p V=n R T .
$$

Here's a more complicated one:

$$
p=\frac{(2 s+1) \sqrt{2 m^3 k_{\mathrm{B}}^5 T^5}}{3 \pi^2 \hbar^3} \int_0^{\infty} \frac{z^{3 / 2} \mathrm{~d} z}{e^{z-\mu /\left(k_{\mathrm{B}} T\right)} \pm 1}
$$

The equations are nonlinear. The xyz components from various equations get multiplied by parameters from other xyz components.

Depending on how much some of the important parameters vary throughout the flow, they may get coupled in other ways in addition to the nonlinearities built in to the equations. For example, some of the "constant" material properties are not actually constant and vary with other parameters. For example. the viscosity of the fluid is dependent on the temperature of the fluid. So that adds more coupling among the equations, and adding further nonlinearity. Sometimes these relationships are modeled as, for example, seventh order polynomials. So now the temperature is tracked up to its 7 th power. Viscosity is also a little dependent on pressure. So just add that into the mix too. So this kind of auxiliary behavior could be thought of as sixth equation.

All of these equations have to be satisfied simultaneously.

In some flows, there can be many orders of magnitude variation in some of the parameters.

When we make enough simplifications, we can end up some versions that can be solved in closed form (i.e. Bernoulli Eqs). One very simplified version is Bernoulli's equation. All of potential flow is based on some tremendous simplifications (constant density, zero viscosity, usually only two dimensions, often steady flow so time derivatives are all zero). Even with these tremendous simplifications, the study of potential flow is very rich and could cover several college level courses.







### Why is the Navier-Stokes equation a millennium prize problem?

In simple terms, the Navier-Stokes equation tells you how a fluid behaves.
A fluid is something that you can assume to be a continuum - i.e. not made of discrete 'Iumps'.

To this end, we can already see that the Navier-Stokes equations must be approximations, because fluids are made of atoms!
However, if you "zoom out" from a fluid, they look pretty much like a continuum. Can you feel the atoms in the air, or does it feel like a smooth continuum fluid?

The Navier-Stokes equations are, in essence, just Newton's 2 nd law written in a form that is applicable to continuum bodies, rather than discrete objects.
Honestly, there's nothing particularly more complex than that in there - at least from a physics perspective.

We know that Newton's second law as $\mathbf{F}=\boldsymbol{m} \mathbf{a}$, but it is more correct to formulate it as:

$$
F=\frac{d}{d t}(m v)
$$

Where $\mathbf{v}$ is the velocity. This obviously simplifies to $\mathbf{F}=\boldsymbol{m} \mathbf{a}$ if the mass is constant.
If we are dealing with a fluid, we don't care about mass, we care about density â€” which for now we assume is constant - i.e., the fluid is incompressible.
Therefore, just by taking the continuum form of Newton's Second Law, we get:

$$
\mathbf{f}=\rho \frac{\mathbf{d v}}{\mathrm{d} t}
$$

Where $\mathbf{f}$ is the "body force" - the infinitesimal force on an infinitesimal chunk of the fluid.
Since the velocity of the fluid is (presumably) both a function of space and of time - we have:

$$
\frac{\mathrm{d}}{\mathrm{d} t} \mathbf{v}(\mathbf{x}, t)=\frac{\partial \mathbf{v}}{\partial t} \frac{\mathrm{d} t}{\mathrm{~d} t}+\frac{\partial \mathbf{v}}{\partial x} \frac{\mathrm{d} x}{\mathrm{~d} t}+\frac{\partial \mathbf{v}}{\partial y} \frac{\mathrm{d} y}{\mathrm{~d} t}+\frac{\partial \mathbf{v}}{\partial z} \frac{\mathrm{d} z}{\mathrm{~d} t}
$$

We note that $\frac{\mathrm{d} x}{\mathrm{~d} t}$ are just components of the velocity, and hence we can put this into a lovely vector notation:

$$
\frac{\mathbf{d}}{\mathrm{d} t} \mathbf{v}(\mathbf{x}, t)=\frac{\partial v}{\partial t}+(\mathbf{v} \cdot \nabla) \mathbf{v}
$$

Which means:

$$
\frac{1}{\rho} \mathbf{f}=\frac{\partial v}{\partial t}+(\mathbf{v} \cdot \nabla) \mathbf{v}
$$

So now we need to think about what these forces could be.
There are two types of forces that can act on a fluid:
- Internal forces, due to interactions between the components
- External forces such as gravity
There are two major internal forces that we need to consider: pressure forces, and the viscosity.
If we write this out, we have:

$$
\mathbf{f}=\mathbf{f}_{v i s c o u s}+\mathbf{f}_{\text {ext }}-\nabla P
$$

Where $\boldsymbol{P}$ is the internal pressure, and $\nabla$ is just a fancy vector derivative.
We plug this in:

$$
\frac{\partial v}{\partial t}+(\mathbf{v} \cdot \nabla) \mathbf{v}=\frac{1}{\rho}\left(\mathbf{f}_{v i s c o u s}+\mathbf{f}_{e x t}-\nabla P\right)
$$

This is the Navier-Stokes equation.
Well, almost - the Navier-Stokes equation has one final step which is to write down the form of the viscous force.

The viscous force is the internal friction of all the particles rubbing up against each other, and takes the form:

$$
\frac{1}{\rho} \mathbf{f}_{v i s c o u s}=\nu \nabla^2 \mathbf{v}
$$

So, we can see that just by applying Newton's 2 nd law to a small chunk of a fluid, we have derived:

$$
\frac{\partial v}{\partial t}+(\mathbf{v} \cdot \nabla) \mathbf{v}=-\frac{1}{\rho} \nabla P+\nu \nabla^2 \mathbf{v}+\frac{1}{\rho} \mathbf{f}_{\text {ext }}
$$

This really is the incompressible Navier-Stokes equation!
So that's what it is, in relatively simple terms: it is just "force is mass times acceleration" souped up a bit, with pressure and viscosity added.

There's honestly nothing more to the physics of it than that.

The mathematics behind this, however, is where things get interesting.

That's because, basically: this is an awful equation to solve.

You can solve it exactly for certain specific setups, and you can make approximations to make it easier to study, but fundamentally....this is just awful.

It's so awful that it hasn't even been proved that solutions always exist.

This is what the Millennium Prize is about - if you are able to show that that equation up there has smooth solutions that exist for any given setup, you will win $\$ 1,000,000$.

They don't even ask you to find the solutions - it's simply showing that they exist that is the problem!

The exact statement is [xx1] :

> Prove or give a counter-example of the following statement:
In three space dimensions and time, given an initial velocity field, there exists a vector velocity and a scalar pressure field, which are both smooth and globally defined, that solve the Navier-Stokes equations.

And don't forget: we have this is only using the equations where it can be assumed that $\rho$ is a constant! The versions of this equations where $\rho$ can vary are...even worse. There's not even $\$ 1,000,000$ on offer regarding that equation, because it's so far out of reach!

So the Navier-Stokes equations are, physically, very simple.

All you are doing is applying classical physics to little chunks of a fluid, and asking how they respond when acted on with a force.

The difficulty is in finding valid mathematical solutions to this.

At this point, the vast majority of interest is in the actual NS equations for mathematical reasons, rather than physical ones.

Whilst physicists make use of them (mostly approximations in the correct regimes), it is mathematicians who will do the heavy lifting.

Physicists are much, much more interested in some of the weird and wonderful behaviour that emerges from these very simple rules - a complex and fascinating subject in its own right.

### What is the relationship between the Navier-Stokes equation and the Schrodinger equation?
They are both partial differential equations that describe the evolution of a system in time.
The Navier-Stokes equation describes fluid flow in time, and the Schrodinger equation describes the evolution of a wave function in time.

Other PDEs that are similar would be the advection, diffusion/heat and wave equations.
